```python
q1=df['Sales'].quantile(0.25)# Calculate the first quartile (Q1)
q3=df['Sales'].quantile(0.75)# Calculate the third quartile (Q3)
iqr=q3-q1# Calculate the interquartile range (IQR)
print(q1, q3, iqr)
```

```python
lower_bound=q1-1.5*iqr# Calculate the lower bound for outlier
```

```python
lower_bound
```

```python
upper_bound = q3 + 1.5 * iqr# Calculate the upper bound for outlier
```

```python
upper_bound
```

```python
required=df[(df['Sales']>=lower_bound)& (df['Sales']<=upper_bound)]
print('All Sales\n', df['Sales'].values)
print('Sales after removing outliers\n', required['Sales'].values)#filtering sales in outliers
```
