### Grouping and Aggregation


```python
grouped = df.groupby("department")
```

```python
grouped
```

```python
df.groupby('department')['salary'].sum()#getting sum of salaries by department
```

```python
df.groupby('department')['salary'].mean()#getting mean of salaries by department
```

```python
df.groupby('department').agg({
'salary': ['sum', 'mean', 'count', 'min', 'max', 'std', 'var', 'median'],
}) #Grouping salary by department
```

```python
df.groupby('department').agg({
'age': ['sum', 'mean', 'count', 'min', 'max', 'std', 'var', 'median'],
})#Grouping age by department
```
